```python
#Mounting Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Memanggil data
# Data dimuat dari file Excel di Google Drive menggunakan pandas dengan path yang ditentukan.
import pandas as pd

file_path = '/content/drive/MyDrive/CAD_alizadeh.xls'
data = pd.read_excel(file_path)
```

```python
# Menampilkan Data
pd.set_option('display.max_rows', None)  # Menampilkan semua baris
pd.set_option('display.max_columns', None)  # Menampilkan semua kolom
pd.set_option('display.width', None)  # Menyesuaikan lebar tampilan
pd.set_option('display.max_colwidth', None)  # Menampilkan teks penuh di kolom
pd.set_option('display.expand_frame_repr', False) # Menampilkan semua data tanpa baris baru

print("Semua data dari dataset:")
print(data)
```

```python
# Preprocessing data: Memeriksa nilai kosong (NaN) di setiap kolom untuk mendeteksi data yang hilang.
print("\nMemeriksa data yang kosong per kolom:")
print(data.isnull().sum())
```

```python
# Pengecekan jenis tiap data untuk diubah ke numerik
print("\nTipe data awal:")
print(data.dtypes)
```

```python
# Mengubah semua kolom menjadi numerik
# menggunakan global replace dengan dictionary mapping
mapping = {
    'Cad' : 0, 'Normal' : 1,
    'Male' : 0, 'Fmale' : 1,
    'Y': 1, 'N': 0,
    'mild': 0.33, 'Moderate': 0.66, 'Severe': 1
}

data = data.replace(mapping)
print(data)
```

```python
# Melakukan proses normalisasi pada masing-masing atribut
from sklearn.preprocessing import MinMaxScaler

# Inisialisasi MinMaxScaler
scaler = MinMaxScaler()

numeric_cols = data.select_dtypes(include=['number']).columns
data_numeric = data[numeric_cols]

# Normalisasi data dan konversi ke DataFrame
data_normalized = pd.DataFrame(scaler.fit_transform(data_numeric), columns=numeric_cols, index=data.index)
data_normalized = data.drop(columns=numeric_cols).join(data_normalized)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
print("\nData setelah normalisasi (semua data):")
print(data_normalized)
```
